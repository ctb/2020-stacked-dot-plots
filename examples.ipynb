```python
import matplotlib.pyplot as plt

import importlib
import alignplot
importlib.reload(alignplot)

from alignplot import AlignmentContainer, StackedDotPlot, AlignmentSlopeDiagram

def show_alignment(A):
    dotplot = StackedDotPlot(A)
    dotplot.plot()

    slope = AlignmentSlopeDiagram(A)
    fig = slope.plot()
    fig.set_size_inches(10,8)

    fig = slope.plot(plot_all_contigs=True, use_labels=False)
    fig.set_size_inches(10,8)
```

## a contaminated genbank genome as query (y axis)

```python
alignment = AlignmentContainer('GCA_003220225.1',
                             ['GCA_003222275.1',],
                             'list.csv', './genomes')
alignment.run_mashmap()
show_alignment(alignment)
```

## TARA genome contaminated with genbank genome

```python
alignment = AlignmentContainer('TARA_ANW_MAG_00083', ('GCA_002171995.1',), 'list.csv', './genomes')
alignment.run_mashmap()
show_alignment(alignment)
```

```python
alignment = AlignmentContainer('GCA_003222275.1', ('GCA_003222275.1',),
                               'list.csv', './genomes')
alignment.run_mashmap()
show_alignment(alignment)
```
